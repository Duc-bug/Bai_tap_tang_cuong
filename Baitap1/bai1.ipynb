```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as ndi
import imageio.v2 as iio
data=iio.imread("kiwi.jpg")
bdata=ndi.shift(data,(30,50,0),mode='wrap')
rows, cols, ch = bdata.shape
X, Y = np.meshgrid(np.arange(cols), np.arange(rows))

offset = 20 * np.sin(2 * np.pi * X / 100)
indices = np.array([Y + offset, X])

wave_img = np.zeros_like(bdata)
for i in range(ch):
    wave_img[..., i] = ndi.map_coordinates(bdata[..., i], indices, order=1, mode='reflect')

iio.imwrite("kiwi_wave.jpg", wave_img.astype(np.uint8))
plt.imshow(data)
plt.title("Hnh gc")
plt.show()

plt.imshow(wave_img)
plt.title("Hnh chnh sa")
plt.show()
```
